```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv('corona_dataset.csv')
data.head()
```

```python
data = data.dropna(axis=0,how="any")
le=LabelEncoder()
data['cardiological pressure']=data['cardiological pressure'].map({'Normal':0,'Elevated':1,'Stage-01':2,'Stage-02':3})
data['Pulmonary_score']=data['Pulmonary_score'].map({'<100':0,'<200':1,'<300':2,'<400':3})
data["Mode_transport"]=le.fit_transform(data["Mode_transport"])
data["Gender"]=le.fit_transform(data["Gender"])
```

```python
data.head()
```

```python
""" 
perfrom data preprocessing on 'data' Dataframe
The final dataframe must be named data
eg .... data =  {preprocessed data}    
"""

data=data.drop(["people_ID","Designation","Name","Married","Gender","comorbidity","Occupation","Pulmonary_score","cardiological pressure","Heart rate","Blood Glucose","FT/month"],axis=1)

onehot=pd.get_dummies(data['Mode_transport'])
data=data.join(onehot)
data=data.drop('Mode_transport',axis=1)

onehot1=pd.get_dummies(data['Region'])
data=data.join(onehot1)
data=data.drop('Region',axis=1)

data
```
