```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

## Data Preprocessing

```python
data = pd.read_csv('corona_dataset.csv')
data.head()
```

```python
data = data.dropna(axis=0,how="any")
le=LabelEncoder()

#LABEL ENCODING
data['cardiological pressure']=data['cardiological pressure'].map({'Normal':0,'Elevated':1,'Stage-01':2,'Stage-02':3})
data['Pulmonary_score']=data['Pulmonary_score'].map({'<100':0,'<200':1,'<300':2,'<400':3})
data["Mode_transport"]=le.fit_transform(data["Mode_transport"])
data["Gender"]=le.fit_transform(data["Gender"])
```

```python
data.corr()
```

```python
data.head()
```

```python
""" 
perfrom data preprocessing on 'data' Dataframe
The final dataframe must be named data
eg .... data =  {preprocessed data}    
"""

# AS FROM THE ABOVE COORELATION WE CAN SEE THAT CERTAIN DATAS ARE LESS SIGNIFICANT TO Infect_Prob SO WE ARE DISCARDING THOSE COLUMNS

data = data.dropna(axis=0,how="any")
data=data.drop(["Designation","Name","people_ID","Married","Gender","comorbidity","Occupation","Pulmonary_score","cardiological pressure","Heart rate","Blood Glucose","FT/month"],axis=1)


#label encoding
"""data['cardiological pressure']=data['cardiological pressure'].map({'Normal':0,'Elevated':1,'Stage-01':2,'Stage-02':3})
data['Pulmonary_score']=data['Pulmonary_score'].map({'<100':0,'<200':1,'<300':2,'<400':3})"""

#onehot encoding
onehot=pd.get_dummies(data['Mode_transport'])
data=data.join(onehot)
data=data.drop('Mode_transport',axis=1)

onehot1=pd.get_dummies(data['Region'])
data=data.join(onehot1)
data=data.drop('Region',axis=1)
data

```

```python
""" separating features and target """

Y = data['Infect_Prob']
X = data.drop(['Infect_Prob'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_X,test_X,train_Y,test_Y = train_test_split(X,Y)

print("Training Size : ",train_X.shape)
print("Testing Size : ",test_X.shape)
```

### Regression to predict the probability of Corona Infection

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
```

```python
""" regression works with numeric values,
make sure you map all the string values in the data to some numeric value if required
"""

regressor.fit(train_X,train_Y)
```

### Performance evaluation

```python
""" complete the function RMSE 
    
    The Root Mean Square Error is defined in the following way
    sqrt(summation((y_pred[i]-y_true[i])^2)/N)
    where:
         i varies from 0 to N
         
    (use numpy)     

"""

def RMSE(y_pred,y_true):
    
    rmse = 0.0
    s=np.subtract(y_true,y_pred)
    l=len(s)
    rmse=(((sum(s*s))/l)**(0.5))
    
    return rmse
    
    

y_pred = regressor.predict(test_X)

print("Root Mean Square Error = ",RMSE(y_pred,test_Y))
```

```python
#Use matplotlib to visualize the total number of cases for different occupations.

dataset=pd.read_csv('corona_dataset.csv')
dataset=dataset.dropna(axis=0,how='any')


#list of occupation is created
occupation=list(dataset["Occupation"].unique())


infected=[]
#A list of infected people is created 
for i in occupation:
    infected.append(sum((dataset['Infect_Prob']>75) & (dataset['Occupation']==i)))
infected
```

```python
import matplotlib.pyplot as plt
```

```python
y=np.arange(len(occupation))
plt.bar(y,infected,align='center',alpha=0.5)
plt.xticks(y,occupation,rotation=70)
plt.xlabel("occupation")
plt.ylabel("No. of people with infected prob >75%")
plt.title("Occupation vs Total Infected People")
plt.show()
```
